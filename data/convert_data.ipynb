# converter

python file to convert from a csv to a json file

```python
import csv
import json

data = []
file_name = ['filosofos', 'materias', 'fontes', 'avaliacoes']

for name in file_name:

    with open(name+'.csv', mode='r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            data.append(row)

    with open(name+'.json', mode='w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```
